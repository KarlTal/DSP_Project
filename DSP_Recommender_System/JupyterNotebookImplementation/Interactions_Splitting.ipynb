```python
import pandas as pd

my_books = pd.read_csv("karl_good_reads.csv", index_col=0)
my_books["book_id"] = my_books["book_id"].astype(str)
```

```python
# As the original csv contains over 229 million user ratings/interactions
# Laptop does not contain enough RAM to go through each and every record to provide
# a recommendation and will take longer than 1 min and was tested to not even 
# complete after 30 mins of non-stop execution.

# Hence, only the first 10 million interactions will be used for the recommendation
# system, this is achieved through pandas, chunking functionality.
# Export into a csv named chunk0.csv
source_path = "goodreads_interactions.csv"
for i,chunk in enumerate(pd.read_csv(source_path, chunksize=10000000)):
    chunk.to_csv('chunk{}.csv'.format(i), index=False)
```
